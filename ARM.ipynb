```python
import importlib
from matplotlib import pyplot as plt
import os
importlib.reload(os)

import mysklearn.mypytable
importlib.reload(mysklearn.mypytable)
from mysklearn.mypytable import MyPyTable

import mysklearn.myutils as myutils
importlib.reload(myutils)

import mysklearn.plot_utils as plt_utils
importlib.reload(plt_utils)

import mysklearn.myclassifiers
importlib.reload(mysklearn.myclassifiers)
from mysklearn.myclassifiers import MyAssociationRuleMiner

data_name = os.path.join("input_data", "322-Project-Table.csv")
projectData = MyPyTable().load_from_file(data_name)
```

```python
before_or_during = projectData.get_column("before-or-during-Covid")

new_usa_increase = projectData.get_column("new-usa-cases-increase-(%)")
ethan = []
for i in range(0, len(new_usa_increase)):
        if new_usa_increase[i] > 0:
            ethan.append("covid-up-TRUE")
        else:
            ethan.append("covid-up-FALSE")

average_airline_increase = projectData.get_column("Average-Airline-Increase-Went-Up?")
ryan = []
for i in range(0, len(average_airline_increase)):
        if average_airline_increase[i] == "TRUE":
            ryan.append("average-airline-up-True")
        else:
            ryan.append("average-airline-up-False")

Xom_went_up = projectData.get_column("XOM-Went-Up?")
adam = []
for i in range(0, len(Xom_went_up)):
        if Xom_went_up[i] == "TRUE":
            adam.append("xom-went-up-True")
        else:
            adam.append("xom-went-up-False")

JBLU_went_up = projectData.get_column("JBLU-Went-Up?")
annemarie = []
for i in range(0, len(Xom_went_up)):
        if JBLU_went_up[i] == "TRUE":
            annemarie.append("jblu-went-up-True")
        else:
            annemarie.append("jblu-went-up-False")

new_table= []
for i in range(0, len(Xom_went_up)):
    if before_or_during[i] == "during":
        new_table.append([ethan[i], ryan[i], adam[i], annemarie[i]])

donkey = MyAssociationRuleMiner()
donkey.__init__(0.1,0.6)
donkey.print_association_rules(new_table)

```
